```python
# No libraries for math functions
def exp_pdf(x, lambd):
    e = 2.71828
    return lambd * (e ** (-lambd * x))

def exp_cdf(x, lambd):
    e = 2.71828
    return 1 - (e ** (-lambd * x))

# Given values
x = 1
lambd = 0.5

pdf_val = exp_pdf(x, lambd)
cdf_val = exp_cdf(x, lambd)

print(f"PDF at x = {x}: {pdf_val}")
print(f"CDF at x = {x}: {cdf_val}")
```

```python
import matplotlib.pyplot as plt

# Values for graph
x_vals = [i * 0.1 for i in range(0, 50)]  # x from 0 to 5
lambd = 0.5
e = 2.71828

pdf_vals = [lambd * (e ** (-lambd * x)) for x in x_vals]
cdf_vals = [1 - (e ** (-lambd * x)) for x in x_vals]

# Plot PDF
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x_vals, pdf_vals, label='PDF', color='blue')
plt.title("Exponential Distribution - PDF")
plt.xlabel("Time (x)")
plt.ylabel("Probability Density")
plt.grid(True)

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_vals, cdf_vals, label='CDF', color='green')
plt.title("Exponential Distribution - CDF")
plt.xlabel("Time (x)")
plt.ylabel("Cumulative Probability")
plt.grid(True)

plt.tight_layout()
plt.show()
```
